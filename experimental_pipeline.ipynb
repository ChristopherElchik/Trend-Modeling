```python
# Import dependencies
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# define constants
RELATIVE_DATA_PATH = 'data/socialmediadata-social-circles-from-facebook/original/facebook/' # start of filepath to data files
```

```python
# testing: read in a graph as an edge list
edgelist_filename = '3980.edges'
digraph = nx.DiGraph()
graph = nx.read_edgelist(path=f"{RELATIVE_DATA_PATH}{edgelist_filename}", create_using=digraph)
list(graph.edges)
```

```python
# visualization experiment
nx.draw(graph, node_size=50, arrows=True, pos=nx.spring_layout(graph))
```

```python
# generate a random watts-strogatz small world and draw it
small_world_graph = nx.watts_strogatz_graph(n=20, k=4, p=0.1, seed=10)
nx.draw(small_world_graph, node_size=50)
```
